```python
# 使用例（複数区間を指定）
input_mp4 = "/home/work/assets/20250308_140538.mp4"
output_mp4 = "merged_output.mp4"

intervals = [
    ("00:00:30", "00:01:00"),  # 1つ目の区間（HH:MM:SS形式または秒数指定）
    ("00:02:00", "00:02:30"),  # 2つ目の区間
    ("00:05:10", "00:05:40")   # 3つ目の区間
]
```

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips

def clip_and_merge(input_file, intervals, output_file):
    clips = []
    with VideoFileClip(input_file) as video:
        for start, end in intervals:
            clip = video.subclip(start, end)
            clips.append(clip)
        final_clip = concatenate_videoclips(clips)
        final_clip.write_videofile(output_file, codec='libx264', audio_codec='aac')
```

```python
clip_and_merge(input_mp4, intervals, output_mp4)
```
